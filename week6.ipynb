# Rolling Windows

This operations are another important transformation for time series data. Rolling windows split the data into time windows and the data in each window is aggregated with a function such as mean aka. moving average

```python
df_7d = df['Close'].rolling(7, center=True).mean()
df_7d.head()
```

Create moving average (SMA) of 10, 25, 50

```python
df_10d = df['Close'].rolling(10, center=True).mean()
df_25d = df['Close'].rolling(25, center=True).mean()
df_50d = df['Close'].rolling(50, center=True).mean()
```

## Preprocessing

Load dataset and perform preprocessing

```python
import pandas as pd
df = pd.read_table('titanic.txt')
df.drop('Name',1, inplace = True)
df.drop('Ticket Number',1,inplace = True)
df.drop('Cabin',1, inplace = True)
df.dtypes
```

Remove rows having missing 'Age' values. 
You must use df.index[ ] to access each rows becuase index doesn't set

```python
df.drop(df.index[df['Age'].isnull()], inplace = True)
```

Remove other missing values

```python
df.dropna(1, inplace = True)
df.dropna(0, inplace = True)
df.dtypes
```

Divide dataset into features data and label data

```python
X = df.iloc[:,0:-1].values
y = df.iloc[:,-1].values
```

Divide dataset into 80% training and 20% test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
X_test.shape
```

Use OrdinalEncoder to convert String to Numerical Data

```python
from sklearn.preprocessing import OrdinalEncoder
enc = OrdinalEncoder()
enc.fit(X_train[:,[0,1]])
X_train[:,[0,1]] = enc.transform(X_train[:,[0,1]])
X_test[:,[0,1]] = enc.transform(X_test[:,[0,1]])
```

Convert data type from object to float

```python
X_train = X_train.astype('float')
X_test = X_test.astype('float')
X_train
```

Perform min-max normalization

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(X_train)
X_train_norm = scaler.transform(X_train)
X_test_norm = scaler.transform(X_test)
X_test_norm
```

Features Reduction using Principle Component Analysis

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca.fit(X_train_norm)
X_reduced = pca.transform(X_train_norm)
X_reduced
```

PCA can help in data visualization

```python
fig, ax = plt.subplots()
ax.scatter(X_reduced[y_train=='Yes',0],X_reduced[y_train=='Yes',1],color='navy',alpha = 0.5)
ax.scatter(X_reduced[y_train=='No',0],X_reduced[y_train=='No',1],color='red',alpha = 0.5)
```

## Modelling

Training to create Decision Tree Model

```python
from sklearn import tree
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train_norm, y_train)
```

Testing Model

```python
y_predict = clf.predict(X_test_norm)
```

Evaluation Model Performance

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_predict)
```

Evaluation Model using Normalized Confusion Metrics

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predict)
cm = cm.astype('float') / cm.sum(axis=1)
cm
```

Create Naive Bayes Model

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(X_train_norm, y_train)
y_pred = nb.predict(X_test_norm)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_predict)
```

## Model Evaluation by Training Accuracy 

Create Support Vector Machine Model

```python
from sklearn.svm import SVC
svm = SVC(C=1, gamma='auto')
svm.fit(X_train_norm, y_train)
y_pred = svm.predict(X_test_norm)
accuracy_score(y_test, y_predict)
```

Cross Validation Accuracy

Cross-validation is performed by separate trainning data into <i>n</i> sub-training set and sub-test set. Then, create model using sub-training and test model using sub-test to get prediction accuracy. Repeat this process until all part of dataset becoming sub-test set. The cross validation accuracy is the average of all sub-test prediction accuracies.

![cross](https://upload.wikimedia.org/wikipedia/commons/1/1c/K-fold_cross_validation_EN.jpg)

```python
from sklearn.model_selection import cross_val_score
score = cross_val_score(svm, X_train_norm, y_train, cv = 10, scoring = 'accuracy')
score.mean()
```
